```python
# nested while loops
num = 0

while num <= 10:
    print(num)
    inner_num = 0
    while inner_num < 5:
        print(inner_num)
        inner_num = inner_num + 1

    num = num + 1
```

```python
# list comprehension

flavors = ['Vanilla', 'chocolate', 'chocolate fudge', 'strawberry', 'choco swirl']
choco_flavours = []

for fla in flavors:
    if 'choco' in fla:
        choco_flavours.append(fla)

print(choco_flavours)
```

```python
choco_flavours_2 = [fla for fla in flavors if 'choco' in fla]
print(choco_flavours_2)
```

```python
n_lis = [x for x in range(20) if x % 2 ==0]
print(n_lis)
```
