```python
# data types
# Data types are a classification (class object) that specifies which type of value a variable has and what operations can be applied to it
# For example, string is a datatype that is used to classify text, integer is a data type used to classify whole numbers
# 5 main data types
# 1. numeric  2. dictionary  3. boolean  4. set  5. Sequence type

# Mutable vs Immutable

# Mutable: the ability of objects to chage their values
# ex: Lists, dictionary, Sets. these datatypes can be updated

# immutablr: Values cannot be changed, once created, the values of these objects are permanennt
# ex: Numbers, Strings, tupels

# Numeric data types are numbers
# 1. integers  2. float  3. complex

# Boolean data types are either true or false
# they are also subtype of Numeric data, evaluating to 0 or 1
# they are used for comparision, logical operators to determine truth or false of an expression

# sequence data type 
# 1. string  2. list  3. tuples
# sequence data types can be accessed via their index, lists are mutable, where as strings and tuples are immutable

# set data type: Set is an unordered collection of data types that is iterable, mutable and has no duplicate elemenets
# sets are used to compare unique values to other set
# sets cannot be accessed by index

# dictionary: items are presented in key:value pairs which makes them unique from every other datatypes
# values in a dictionary can be of any datatypes and can be duplicated, whereas keys cannot be repeated and must be immutable 


```
