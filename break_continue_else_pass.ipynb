```python
# while loops: we use while loop when we dont know how many times we want to iterate before hand
# for loop iterates over a sequence and while loop iterates over a block of code

lists = ['var', 'par', 'kar', 'aar']

i=0

while lists[i] != 'kar':
    print(lists[i])
    i=i+1
```

```python
# break, continue, else, pass are used to control our logic in while loops
# break: break is used to terminate a while loop or for loop. statements after break will not execute

counter = 0

while counter <= 10:
    print(counter)
    counter = counter+1
```

```python
counter = 0

while counter <= 10:
    print(counter)
    if counter == 6:
        break
    counter = counter+1
```

```python
counter = 0

while counter <= 10:
    if counter == 6:
        break
    print(counter)
    counter = counter+1
```

```python
ints = [1,2,3,4,5]

for num in ints:
    pow_of_num = num ** num
    if pow_of_num > 2500:
        break
    print(pow_of_num)
```

```python
num =0

while num <6:
    num = num +1
    if num == 2:
        continue
    print(num)
```

```python
intss = [1,2,3,4,5,6,7]

for ints in intss:
    power_of_num = ints ** ints
    if 200 <= power_of_num <= 2500:
        continue
    print(power_of_num)
```
