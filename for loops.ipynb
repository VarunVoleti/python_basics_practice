```python
# nested if

grade =85

if grade >= 90:
    if grade >= 97:
        print(str(grade) + ' = A+')
    elif grade >= 93:
        print(str(grade) + ' = A')
    else:
        print(str(grade) + ' = A-')
elif grade >= 80:
    if grade >= 87:
        print(str(grade)+ ' = B+')
    elif grade >= 85:
        print(str(grade)+ ' = B')
    else:
        print(str(grade)+ ' =B-')
```

```python
sale_amount = [1,10,15,6,80,54,29,44,37,97,57,98,76]

small_sale = []
medium_sale = []
large_sale = []

for sale in sale_amount:
    if sale < 10:
        small_sale.append(sale)
    elif sale < 50:
        medium_sale.append(sale)
    else:
        large_sale.append(sale)

print('Small sale : ', small_sale)
print('Medium sale : ', medium_sale)
print('Large sale : ', large_sale)
```

```python
names = ['var','par','kar','aar']

i=1 # i is a counter

for name in names:
    print(i, name)
    i= i+1
```

```python
for a, name in enumerate(names): # enumerate works like natural or default counter
    print(a,name)
```
